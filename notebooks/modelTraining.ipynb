##Training the model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['month', 'year','day'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
#Encodding of Categorical Features
df['Classes']=np.where(df['Classes']=='not fire',0,1)
```

```python
df.head()
```

```python
df['Classes'].value_counts()
```

```python
#Independent and Dependent Features

X = df.drop('FWI', axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
y
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
##Feature Selection based on correlation
X_train.corr()
```

```python
##Checking for Multicollinearity
plt.figure(figsize = (12,10))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
#Threshold--Domain Expertise
corr_features = correlation(X_train,0.85)
```

```python
 ## Drop Feature when correlation is more than threshold
X_train.drop(corr_features,axis=1,inplace=True)
```

```python
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

##Feature Scaling and Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) 
```

```python
X_train_scaled
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("Before Scaling and Standardization")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title("After Scaling and Standardization")
```

##Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
linreg = LinearRegression()
linreg.fit(X_train_scaled,y_train)
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error: ", mae)
print("R2 Score: ", score)
plt.scatter(y_test,y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r', label='Best Fit Line (y=x)')
```

##Lasso Regression Model

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error,r2_score
lasso = Lasso()
lasso.fit(X_train_scaled,y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error: ", mae)
print("R2 Score: ", score)
plt.scatter(y_test,y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r', label='Best Fit Line (y=x)')
```

###Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv=5)
lassocv.fit(X_train_scaled, y_train)
```

```python
lassocv.alphas_
```

```python
lassocv.mse_path_
```

```python
y_pred=lassocv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error: ", mae)
print("R2 Score: ", score)
```

##Ridge Regression Model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error,r2_score
ridge = Ridge()
ridge.fit(X_train_scaled,y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error: ", mae)
print("R2 Score: ", score)
plt.scatter(y_test,y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r', label='Best Fit Line (y=x)')
```

###Ridge Regression CV Model

```python
from sklearn.linear_model import RidgeCV
ridgecv = RidgeCV(cv=5)
ridgecv.fit(X_train_scaled,y_train)
y_pred=ridgecv.predict(X_test_scaled)
plt.scatter(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
print("Mean Absolute Error: ", mae)
print("R2 Score: ", score)
```
