##Algerian Forest Fires

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv", header=1)
dataset.head()
```

```python
dataset.info()
```

##Data Cleaning

```python
#missing values
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df=dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Dropping the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Fix spaces in the column
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
##Changing the required columns to required data type

df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']]=df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype(float)
```

```python
df.info()
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.head()
```

```python
##Save the cleaned dataset 

df.to_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv', index=False)
```

##Exploratory Data Analysis

```python
df_copy = df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
#Encodding of Categorical Features
df_copy['Classes']=np.where(df_copy['Classes']=='not fire',0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
df_copy.tail()
```

```python
df_copy = df_copy.drop('CLases', axis=1)
```

```python
df.info()
```

```python
df_copy.head()
```

```python
#Plot density for all features

plt.style.use('ggplot')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
##Percentage of Pie Chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
classlabel = ['Fire','Not Fire']
plt.figure(figsize=(25,5))
plt.pie(percentage,labels=classlabel, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
##Correlation 
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr(), annot=True)
```

```python
##Box Plot for Outlier

sns.boxplot(df['FWI'], color='red')
```

```python
## Monthly Fire Analysis
df_temp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes', data=df)
plt.ylabel('No of Fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Sidi-Bel Abess Fires')
```
